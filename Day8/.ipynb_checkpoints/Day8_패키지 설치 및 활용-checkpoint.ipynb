### Apply(Ìï®Ïàò)
* `.apply()`
* Ïñ¥Îñ§ Ìï®ÏàòÎ•º series Îç∞Ïù¥ÌÑ∞Ïùò Î™®Îëê Îã§ Ï†ÅÏö©ÌïòÍ≥† Ïã∂ÏùÑ Îïå ÏÇ¨Ïö©
    * index ÌïòÎÇòÏùò Î™®Îì† Îç∞Ïù¥ÌÑ∞Î•º Î≥ÄÍ≤Ω or column ÌïòÎÇòÏùò Î™®Îì† Îç∞Ïù¥ÌÑ∞Î•º Î≥ÄÍ≤Ω
* ÏÇ¨Ïö©ÌïòÍ∏∞ ÏúÑÌï¥ÏÑú Î®ºÏ†Ä Ìï®ÏàòÎ•º ÏÑ†Ïñ∏Ìï¥ÏïºÌï®.

```python
def isPass(t):
    if t>20:
        return 'Pass'
    else:
        return 'Fail'
```

```python
#.apply(ÏÑ†Ïñ∏Ìïú Ìï®Ïàò)
mpg['total'].apply(isPass)
```

```python
# ÏÉàÎ°úÏö¥ ÌååÏÉùÎ≥ÄÏàò + Apply
mpg['test'] = mpg['total'].apply(isPass)
```

```python
# Í≤∞Í≥º Í∞í ÌôïÏù∏
mpg['test'].value_counts()
```

```python
import pandas as pd # Ìå®ÌÇ§ÏßÄ Î∂àÎü¨Ïò§Í∏∞
```

```python
# Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞
people_df = pd.read_csv('midwest.csv')
```

```python
# 1Î≤à
# info
people_df.info()
```

```python
# ÏùºÎ∂Ä ÌôïÏù∏
people_df.head()
```

```python
# 2Î≤à
# ÎÇ¥Ïö© Î≥ÄÍ≤Ω (ÏÉàÎ°úÏö¥ columns Ï∂îÍ∞Ä Î∞è Í∏∞Ï°¥ columns Ï†úÍ±∞)
people_df['total'] = people_df['poptotal']
people_df.drop('poptotal',axis =1, inplace=True)
people_df['asian'] = people_df['popasian']
people_df.drop('popasian', axis =1, inplace=True)
# rename ÏÇ¨Ïö©ÎèÑ Í∞ÄÎä•.
# midwest.rename({'poptotal': 'total', 'popasian': 'asian'}, axis = 1, inplace = True)
# ÌôïÏù∏
people_df.info()
```

```python
# 3Î≤à
people_df['Asian_percent'] = people_df['asian'] / people_df['total'] * 100
```

```python
# ÌôïÏù∏
people_df['Asian_percent'].value_counts
```

```python
# 4Î≤à
import numpy as np
asian_mean = people_df['Asian_percent'].mean()
# ÌôïÏù∏
asian_mean
```

```python
# Î∂ÑÎ•òÌï¥ÏÑú ÏÑ†Ïñ∏
asian_per = np.where(people_df['Asian_percent'] > asian_mean, 'large',
                             'small')
# Í∞í ÎåÄÏûÖ
people_df['Large & Small'] = asian_per
# 
```

```python
people_df['Large & Small'].value_counts()
```

```python
# 4Î≤à applyÏÇ¨Ïö©
# Ìï®Ïàò ÏÑ†Ïñ∏
def isSize(percent, mean):
    if percent > mean:
        return 'large'
    return 'small'
```

```python
# apply ÏÇ¨Ïö©
# mapÍ≥º ÎπÑÏä∑
# args ÏÇ¨Ïö© : Ïù∏ÏûêÎ•º ÎÑòÍ≤®Ï£ºÍ∏∞ ÏúÑÌï¥ÏÑú ÏÇ¨Ïö©(Ï°∞ÏÇ¨ ÌïÑÏöî)
people_df['l & s'] = people_df['Asian_percent'].apply(isSize, args=[people_df['Asian_percent'].mean()])
```

```python
# ÌôïÏù∏
people_df['l & s'].value_counts()
```

```python
import pandas as pd
```

```python
# data Î∂àÎü¨Ïò§Í∏∞
people = pd.read_csv('midwest.csv')
# data ÌôïÏù∏
people.info()
```

```python
# ÏùºÎ∂Ä ÌôïÏù∏
people.head()
```

```python
# columns gÌôïÏù∏ 
people.columns
```

```python
# ÎØ∏ÏÑ±ÎÖÑ Îç∞Ïù¥ÌÑ∞ Ï∂îÍ∞Ä
people['popchild'] = people['poptotal'] - people['popadults']
people['popchild']
```

```python
# 1Î≤à
people['perchild'] = people['popchild'] / people['poptotal'] * 100
people['perchild']
```

```python
# 2Î≤à
# sort_valeus Îç∞Ïù¥ÌÑ∞ Ï†ïÎ†¨
hi_pc = people.sort_values('perchild',ascending=False)
hi_pc
```

```python
# ÏÉÅÏúÑ 5Í∞úÎßå Ï∂úÎ†•
# headÎèÑ Í∞ÄÎä•
print(hi_pc[['county','perchild']].head())
print(hi_pc[['county','perchild']].iloc[:5]) # slicing 
print('--------')
hi_pc.loc[:5,['county','perchild']]
```

```python
# 3Î≤à
# numpy ÏÑ†Ïñ∏
import numpy as np
child = np.where(people['perchild'] >= 40, 'large',
        np.where(people['perchild'] >= 30, 'middle',
        'small'))
people['child l & m & s'] = child
# ÌôïÏù∏
people['child l & m & s'].value_counts()
```

```python
# 4Î≤à
people['perasian'] = people['popasian'] / people['poptotal'] * 100
people['perasian']

low_per_asian = people['perasian'].sort_values().head(10)
low_per_asian.values[-1]
```

```python
# Ï†ÑÏ≤¥ Îç∞Ïù¥ÌÑ∞Ïùò 10Í∞ú Í∞í Í∞ÄÏ†∏Ïò§Í∏∞
low_t = people[people['perasian'] <= low_per_asian.values[-1]]
low_t
```

```python
# state, country, Î∞±Î∂ÑÏú® Ï∂úÎ†•
low_t[['county','state','perasian']]
# Í∞ôÏùÄ Î∞©Î≤ï
# low_t.loc[:,['county','state','perasian']]
```

```python
# ÏúÑÏóêÏÑú stateÍ∞Ä Í≤πÏπòÎØÄÎ°ú Ïù¥Ïóê ÎåÄÌï¥ÏÑú Î∂ÑÎ•òÌïúÎã§.
# stateÎ•º Í∏∞Ï§ÄÏúºÎ°ú Î∂ÑÎ•ò -> Í∑∏ Ï§ë perasianÏùò ÌèâÍ∑†ÏùÑ Íµ¨ÌïòÍ≥† -> ÌÅ¨Í∏∞ÎåÄÎ°ú Ï†ïÎ†¨
people.groupby('state')['perasian'].mean().sort_values()
```
