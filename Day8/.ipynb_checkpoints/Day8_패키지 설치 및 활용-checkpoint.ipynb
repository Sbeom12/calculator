#### 얕은 복사 & 깊은 복사
* df변수명
*  `df2 = df2`
    * 얕은 복사(shallow copy)'
    * 원본에 영향이 됨.
* `df3 = df.copy()`
    * 깊은 복사(deep copy)
    * 원본에 영향이 안됨.
    * 같은 데이터가 2개

```python
# 위에서 선언한 데이터들이 다르다는 것을 알 수 있다.
print(id(df))
print(id(df2))
print(id(df3))
```

### Apply(함수)
* `.apply()`
* 어떤 함수를 series 데이터의 모두 다 적용하고 싶을 때 사용
    * index 하나의 모든 데이터를 변경 or column 하나의 모든 데이터를 변경
* 사용하기 위해서 먼저 함수를 선언해야함.

```python
def isPass(t):
    if t>20:
        return 'Pass'
    else:
        return 'Fail'
```

```python
#.apply(선언한 함수)
mpg['total'].apply(isPass)
```

```python
# 새로운 파생변수 + Apply
mpg['test'] = mpg['total'].apply(isPass)
```

```python
# 결과 값 확인
mpg['test'].value_counts()
```

```python
import pandas as pd # 패키지 불러오기
```

```python
# 데이터 불러오기
people_df = pd.read_csv('midwest.csv')
```

```python
# 1번
# info
people_df.info()
```

```python
# 일부 확인
people_df.head()
```

```python
# 2번
# 내용 변경 (새로운 columns 추가 및 기존 columns 제거)
people_df['total'] = people_df['poptotal']
people_df.drop('poptotal',axis =1, inplace=True)
people_df['asian'] = people_df['popasian']
people_df.drop('popasian', axis =1, inplace=True)
# rename 사용도 가능.
# midwest.rename({'poptotal': 'total', 'popasian': 'asian'}, axis = 1, inplace = True)
# 확인
people_df.info()
```

```python
# 3번
people_df['Asian_percent'] = people_df['asian'] / people_df['total'] * 100
```

```python
# 확인
people_df['Asian_percent'].value_counts
```

```python
# 4번
import numpy as np
asian_mean = people_df['Asian_percent'].mean()
# 확인
asian_mean
```

```python
# 분류해서 선언
asian_per = np.where(people_df['Asian_percent'] > asian_mean, 'large',
                             'small')
# 값 대입
people_df['Large & Small'] = asian_per
# 
```

```python
people_df['Large & Small'].value_counts()
```

```python
# 4번 apply사용
# 함수 선언
def isSize(percent, mean):
    if percent > mean:
        return 'large'
    return 'small'
```

```python
# apply 사용
# map과 비슷
# args 사용 : 인자를 넘겨주기 위해서 사용(조사 필요)
people_df['l & s'] = people_df['Asian_percent'].apply(isSize, args=[people_df['Asian_percent'].mean()])
```

```python
# 확인
people_df['l & s'].value_counts()
```

```python
import pandas as pd
```

```python
# data 불러오기
people = pd.read_csv('midwest.csv')
# data 확인
people.info()
```

```python
# 일부 확인
people.head()
```

```python
# columns g확인 
people.columns
```

```python
# 미성년 데이터 추가
people['popchild'] = people['poptotal'] - people['popadults']
people['popchild']
```

```python
# 1번
people['perchild'] = people['popchild'] / people['poptotal'] * 100
people['perchild']
```

```python
# 2번
# sort_valeus 데이터 정렬
hi_pc = people.sort_values('perchild',ascending=False)
hi_pc
```

```python
# 상위 5개만 출력
# head도 가능
print(hi_pc[['county','perchild']].head())
print(hi_pc[['county','perchild']].iloc[:5]) # slicing 
print('--------')
hi_pc.loc[:5,['county','perchild']]
```

```python
# 3번
# numpy 선언
import numpy as np
child = np.where(people['perchild'] >= 40, 'large',
        np.where(people['perchild'] >= 30, 'middle',
        'small'))
people['child l & m & s'] = child
# 확인
people['child l & m & s'].value_counts()
```

```python
# 4번
people['perasian'] = people['popasian'] / people['poptotal'] * 100
people['perasian']

low_per_asian = people['perasian'].sort_values().head(10)
low_per_asian.values[-1]
```

```python
# 전체 데이터의 10개 값 가져오기
low_t = people[people['perasian'] <= low_per_asian.values[-1]]
low_t
```

```python
# state, country, 백분율 출력
low_t[['county','state','perasian']]
# 같은 방법
# low_t.loc[:,['county','state','perasian']]
```

```python
# 위에서 state가 겹치므로 이에 대해서 분류한다.
# state를 기준으로 분류 -> 그 중 perasian의 평균을 구하고 -> 크기대로 정렬
people.groupby('state')['perasian'].mean().sort_values()
```
