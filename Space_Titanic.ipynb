# 호진 썜

```python
# 데이터 가져오기
import pandas as pd
import numpy as np
data=pd.read_csv('train.csv')
real=pd.read_csv('test.csv')
submit=pd.read_csv('sample_submission.csv')
train = data.copy()
```

## HomePanet 결측치 처리

```python
train['Group'] = data['PassengerId'].str[:4]
train
```

```python
# size 
home_group=train.groupby(['Group','HomePlanet']).size().reset_index()
home_group
```

```python
# 위와 동일한 방법
train.groupby('Group', as_index = False)['HomePlanet'].value_counts()
```

```python
# 뒤를 위한 일부 확인 하나만 뽑아서 가져오기 
home_group.loc[0, 'Group']
```

```python
# 딕셔너리 형태로 key : Group 데이터, value = HomePLanet 데이터 이용
group_home_dict={}
for i in range(6107):
    group_home_dict[home_group.loc[i,'Group']]=home_group.loc[i,'HomePlanet']
# 확인
group_home_dict
```

```python
# zip으로도 가능
for g,hp in zip(home_group['Group'],home_group['HomePlanet']):
    print(g,hp)
```

```python
# 위에서 선언한 dictionary로 결측치 채우기
train['HomePlanet'].fillna(train['Group'].map(group_home_dict), inplace=True)
```

```python
# 기존 결측치 확인
data['HomePlanet'].isnull().sum()
```

```python
# 새로운 결측치 확인
train['HomePlanet'].isnull().sum()
```

위에서 가족끼리 온 결측치를 처리했지만, 혼자 온 사람이 있거나 같은 group으로 묶이지 않음

```python
# 이름의 성만 가져오기
train['Name'].str.split(' ').str[-1]
```

```python
# 이름의 성만 가져와서 Group2로 선언
group2=[]
for i in train['Name'].str.split(' '):
    try:
        group2.append(i[1])
    except: 
        group2.append(None)
train['Group2'] = group2
train['Group2']
```

```python
# 위와 같은 방법으로 묶어서 사용
home_group2 = train.groupby(['Group2','HomePlanet']).size().reset_index()
home_group2
```
